**Observations**

Même si globalement le nombre d'offres de CDI est toujours bien supérieur aux autres types de contrat (ce qui est logique puisqu'il y a beaucoup plus d'offres de CDI), on peut dégager certaines tendances. 

Par exemple, il y a beaucoup beaucoup (beaucoup) plus d'offres de CDI dans le département 101 correspondant au "Sales business development".

Voyons si le secteur d'activité de l'entreprise (industry_id) peut nous permettre de préciser certaines tendances.

```python

df_offers.groupby(['contract_id','industry_id'])['contract_id'].count()
```

**Observations**

Nous dégageons d'autres tendances intéressantes. Par exemple, nous remarquons que les offres de CDI sont plus nombreuses dans les secteurs 9, 10 et 20 (respectivement bâtiment, consulting, media). Tout comme les offres de V.I.E qui sont plus présentes dans les secteurs 16, 17, 19 (respectivement industrie, technologie, internet).
