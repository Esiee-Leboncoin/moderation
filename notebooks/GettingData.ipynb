```python
from toolbox.db_access.postgres import PSQManagement
```

```python
psqm = PSQManagement()
```

```python
columns = ["id as offer_id", "title", "description", "contract_id", "recruiter_id", "department_id", 
           "industry_id", "location", "latitude", "longitude", "created_at"]
OFFER_NUMBER = 10000
```

```python
df = psqm.get_df_from_query(f"SELECT {', '.join(columns)} FROM offers ORDER BY created_at DESC LIMIT {OFFER_NUMBER};")
```

```python
df.to_csv("./../data/offers.csv", index=False)
```

```python
recruiter_columns = [
    "id as recruiter_id",
    "industry_id",
    "organization_id"
]
df_recruiters = psqm.get_df_from_query(f"SELECT {', '.join(recruiter_columns)} FROM recruiters WHERE id IN ({','.join(df.recruiter_id.astype(str).tolist())}) LIMIT {OFFER_NUMBER};")
```

```python
df_recruiters.to_csv("./../data/recruiters.csv", index=False)
```

```python
organization_columns = [
    "id as organization_id",
    "name",
    "location", 
    "latitude",
    "longitude", 
    "industry_id",
    "description",
    "siret"
]
df_organization = psqm.get_df_from_query(f"SELECT {', '.join(organization_columns)} FROM organizations WHERE id IN ({','.join(df_recruiters.organization_id.astype(str).tolist())}) LIMIT {OFFER_NUMBER};")
```

```python
df_organization.shape
```

```python
df_organization.to_csv("./../data/organizations.csv", index=False)
```
